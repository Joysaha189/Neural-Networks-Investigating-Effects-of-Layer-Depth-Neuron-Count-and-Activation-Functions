```
import matplotlib.pyplot as plt
import numpy as np

# Example data (replace with your actual accuracy values)
datasets = ['F-MNIST', 'MNIST', 'Iris', 'MLL', 'Madelon', 'Leukemia']

# Accuracy values for training
relu_train = [0.9281, 0.9969, 0.50, 1, 0.4985, 0.6666]
leaky_relu_train = [0.9351, 0.9982, 0.70, 0.9824, 0.50144, 0.35087]
sigmoid_train = [0.8307, 0.9130, 1, 1, 0.4985, 0.1578]

# Accuracy values for testing
relu_val = [0.9281, 0.9681, 0.51, 1, 0.5057, 0.649]
leaky_relu_val = [0.9351, 0.9779, 0.78, .8666, 0.4942, 0.333]
sigmoid_val = [0.8307, 0.9082, 0.83, .5333, 0.5057, 0.20]

# Set the position of bars on the x-axis
x = np.arange(len(datasets))

# Set the width of the bars
width = 0.2  # the width of the bars

# Create the bar plots
fig, ax = plt.subplots(figsize=(12, 6))
fig.patch.set_facecolor('lightblue')

# Plotting the bars for each activation function and both training and testing accuracy
bars1_train = ax.bar(x - width, relu_train, width, label='ReLU Train', color='royalblue', hatch='//')
bars2_train = ax.bar(x, leaky_relu_train, width, label='LeakyReLU Train', color='orchid', hatch='\\')
bars3_train = ax.bar(x + width, sigmoid_train, width, label='Sigmoid Train', color='lightgreen', hatch='-')

# Testing accuracy bars (overlaid with transparency)
bars1_test = ax.bar(x - width, relu_val, width, label='ReLU Validation', color='lightcyan', alpha=0.7)
bars2_test = ax.bar(x, leaky_relu_val, width, label='LeakyReLU Validation', color='magenta', alpha=0.7)
bars3_test = ax.bar(x + width, sigmoid_val, width, label='Sigmoid Validation', color='lightgray', alpha=0.7)


# Adding labels and title
ax.set_xlabel('Dataset',fontsize=14)
ax.set_ylabel('Accuracy',fontsize=14)
ax.set_title('Training vs Validation Accuracy for Different Activation Functions',fontsize=18)
ax.set_xticks(x)
ax.grid(True, linestyle='--', alpha=.6, zorder=0, axis='y',linewidth=1,color='black')  # Only show horizontal grid lines
ax.set_xticklabels(datasets, rotation=45)
ax.legend()

# Display the plot
plt.tight_layout()
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np

# Example data
datasets = ['F-MNIST', 'MNIST', 'Iris','MLL', 'Madelon','Leukemia']
accuracies = [0.9351, 0.9779, .986, 1, 0.5057, 0.649]

# Set the position of points on the x-axis
x = np.arange(len(datasets))

# Define different markers and colors for each point
markers = ['o', 's', '^', 'D', 'v', '*']
colors = ['blue', 'green', 'red', 'purple', 'orange', 'cyan']

# Create the scatter plot
fig, ax = plt.subplots(figsize=(10, 6))

# Set the background color
fig.patch.set_facecolor('lightblue')

# Plot each point with a different marker and color
for i in range(len(datasets)):
    ax.scatter(x[i], accuracies[i], color=colors[i], marker=markers[i], s=200,
               label=datasets[i], edgecolors='black', zorder=5)

# Adding grid, labels, and title
ax.grid(True, linestyle='-', axis= 'y',alpha=0.9, zorder=0)  # Set zorder for grid lines

# Change grid background color using axhspan and axvspan
ax.axhspan(0, 1.1, facecolor='white', alpha=0.5, zorder=-1)
ax.set_facecolor("white")
ax.axhspan(0, 1.1, facecolor='white', alpha=0.5, zorder=-1)  # Horizontal span


ax.set_xlabel('Dataset',fontsize=14)
ax.set_ylabel('Accuracy',fontsize=14)
ax.set_title('Best Validation Accuracy for Different Datasets',fontsize=18)
ax.set_xticks(x)
ax.set_xticklabels(datasets, rotation=45, fontsize= 12)

# Adding text labels for each point
for i, accuracy in enumerate(accuracies):
    ax.text(i, accuracy + 0.05, f'{accuracy:.2f}', ha='center', va='bottom', fontsize=15)

# Display the plot
plt.tight_layout()
plt.legend(fontsize= 12,
           markerscale= 1,
           labelspacing= 1,
           loc= 'lower left')
plt.show()
```
